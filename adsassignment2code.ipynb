```python
'''
creating a sub data frame with selected rows and coloumns from the data frame 
'''
Ags = df[df['Series Code']=='AG.LND.AGRI.ZS']
```

```python
'''
Check if each element in the DataFrame is contained in values.
'''
agricultural_land = Ags[Ags['Country Code'].isin(['AUS','BHR','CAN','CHN','ARE','IND'])]
agricultural_land
```

```python
Trans= agricultural_land.transpose()
'''
here we are modifying the dataframe by changing the row elements into column elements and the column elements into row elements.
The output of this function is a modified array of the original one.

'''

Trans.rename(columns=Trans.iloc[0], inplace = True)

agricultural_land=Trans.iloc[1:]
```

```python
# removing the null values from the dataframe
agricultural_land.fillna(0,inplace=True)
```

```python
#print the dataframe
agricultural_land.head(30)
```

```python
'''  Determines the data type of the given data
'''
agricultural_land.dtypes
```

```python
'''
coverting the  Object type data into float type using numeric() function 

'''
agricultural_land['AUS'] = pd.to_numeric(agricultural_land['AUS'],errors = 'coerce')
agricultural_land['CAN'] = pd.to_numeric(agricultural_land['CAN'],errors = 'coerce')
agricultural_land['BHR'] = pd.to_numeric(agricultural_land['BHR'],errors = 'coerce')
agricultural_land['CHN'] = pd.to_numeric(agricultural_land['CHN'],errors = 'coerce')
agricultural_land['ARE'] = pd.to_numeric(agricultural_land['ARE'],errors = 'coerce')
agricultural_land['IND'] = pd.to_numeric(agricultural_land['IND'],errors = 'coerce')
```

```python
'''creating lineplot using given data
'''
ax = agricultural_land[['AUS','CAN','BHR','CHN','ARE','IND']].plot(kind='line' , title="agricultural_land(%)" , figsize=(11,6) , legend=True , fontsize=12)

ax.set_xlabel("Country Name", fontsize=12)
ax.set_ylabel("n", fontsize=12)
plt.show()
```
